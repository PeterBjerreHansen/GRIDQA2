```python
import json
import numpy as np
import matplotlib.pyplot as plt
from generate_dataset import Sample
from formatter import Formatter

# Load dataset
with open("dataset.json", "r") as f:
    dataset = [Sample.model_validate(sample) for sample in json.load(f)]
```

```python
# Count unique
labels = [dat.transformation.type.value for dat in dataset]
values, counts = np.unique(labels, return_counts=True)
print(list(zip(values, counts)))
```

```python
for i in range(100):
    print(dataset[i].transformation)
    before_grid = dataset[i].grid_before
    after_grid = dataset[i].grid_after
    # after_grid = np.rot90(before_grid, 90)

    grid_shape_x, grid_shape_y = np.shape(before_grid)
    fig, axs = plt.subplots(1, 2, figsize=(10, 5))
    axs[0].set_title("Before")
    axs[1].set_title("After")
    # axs[0].set_xticks(np.arange(0, grid_shape_x)+0.5)
    # axs[0].set_yticks(np.arange(0, grid_shape_y)+0.5)
    # axs[1].set_xticks(np.arange(0, grid_shape_x)+0.5)
    # axs[1].set_yticks(np.arange(0, grid_shape_y)+0.5)
    # axs[0].grid(True)
    # axs[1].grid(True)
    axs[0].imshow(before_grid, cmap="Greys", interpolation='none')
    axs[1].imshow(after_grid, cmap="Greys", interpolation='none')
    axs[0].tick_params(labelbottom=False, labelleft=False)
    axs[1].tick_params(labelbottom=False, labelleft=False)
    plt.show()
```

```python
letter_mapping = {
    1: "R",
    2: "G",
    3: "B",
    4: "Y",
    5: "P",
}

formatter = Formatter()
sample = dataset[1]
print(formatter.grid_to_text(np.array(sample.grid_before), letter_mapping))

print(formatter.grid_to_text(np.array(sample.grid_after), letter_mapping))
```
